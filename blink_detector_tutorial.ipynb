```python
import cv2
import imutils
from matplotlib import pyplot as plt
```

```python
img_url = "https://static.thairath.co.th/media/dFQROr7oWzulq5Fa5ya8MgjNF5bFbrVgYmydWLrlCCuPlxEhlpDYsIiTjATAH4N235z.jpg"
img_url_penguin = "https://www.cabq.gov/artsculture/biopark/news/10-cool-facts-about-penguins/@@images/1a36b305-412d-405e-a38b-0947ce6709ba.jpeg"
frame = imutils.url_to_image(img_url)
img_penguin = imutils.url_to_image(img_url_penguin)

fig, axes = plt.subplots(1,2)

for ax, im in zip(axes.ravel(), [frame, img_penguin]):
    ax.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    ax.axis("off")
```

```python
from blink_detector import detect_faces

faces = detect_faces(frame)
print(faces)

no_face = detect_faces(img_penguin)
print(no_face)
```

```python
trackers = cv2.legacy.MultiTracker_create()
```

```python
from blink_utilities import convert_rect_to_bb

frame_detected = frame.copy()

for face in faces:
    x, y, w, h = convert_rect_to_bb(face)
    cv2.rectangle(frame_detected, (x, y), (x+w, y+h), [0,0,255], 2)

plt.imshow(cv2.cvtColor(frame_detected, cv2.COLOR_BGR2RGB))
```
